```python
class subplot: 
    def __init__(self): 
        self.log=logger()
    def plottingcountplot(self): 
        plt.figure(figsize=(15,60))
        plotnumber=1
        for i in phishing.drop(columns='Result',axis=1).columns:
            ax = plt.subplot(12,3,plotnumber)
            self.log.apply_log(file_name='EDA.log',msg='Inside plottingcountplot method and countplot created')
            sns.countplot(data=phishing,x=phishing[i])
            plt.xlabel(i)
            plotnumber=plotnumber+1
        plt.show()
```

```python
test=subplot()
```

```python
test.plottingcountplot()
```

```python
import logging as lg


class logger:
    def __init__(self):
        pass
    def apply_log(self, file_name, msg):
        self.file=file_name
        self.message=msg
        lg.basicConfig(filename=self.file, filemode='a',level=lg.INFO,
                       format='%(asctime)s : %(name)s : %(levelname)s : %(message)s',
                       datefmt='%d/%m/%Y  %I:%M:%S %p')
        lg.info(self.message)
```
